```python
import pandas as pd
import numpy as np
```

```python
df_raw= pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
display(df_raw)
```

```python
df_raw.isnull().sum()
```

#Data Cleaning to eliminate rows with 's' in the column 'Mean Score'

```python
df = df_raw.loc[~df_raw['Mean Score'].str.contains('s')]
display(df)
```

```python
df_Queens= df.loc[df['School DBN'].str.contains('Q')]
display(df_Queens)
```

```python
df_Brooklyn= df.loc[df['School DBN'].str.contains('K')]
display(df_Brooklyn)
```

#I want to only show the schools labelled as "High School" in Brooklyn




```python
df_Brooklyn_HS= df_Brooklyn.loc[df_Brooklyn['School Level'].str.contains('High school')]
display(df_Brooklyn_HS)
```

#I want to only show the schools labelled as "High School" in Queens


```python
df_Queens_HS= df_Queens.loc[df_Queens['School Level'].str.contains('High school')]
display(df_Queens_HS)
```

#Limit the High schools in brooklyn - by regents exam 'Common Core Algebra'




```python
df_B_HS_CCA= df_Brooklyn_HS.loc[df_Brooklyn_HS['Regents Exam']==('Common Core Algebra')]
display(df_B_HS_CCA)
```

#Limit the High schools in queens - by regents exam 'Common Core Algebra'


```python
df_Q_HS_CCA= df_Queens_HS.loc[df_Queens_HS['Regents Exam']==('Common Core Algebra')]
display(df_Q_HS_CCA)
```

#I picked the school PS 350K District 13 in Brooklyn (School DBN is 13K350)

```python
df_13K350 = df_B_HS_CCA.loc[df_B_HS_CCA['School DBN'].str.contains('13K350')]
display(df_13K350)
```

```python
df_13K350_MeanScore = df_13K350['Mean Score']
df_13K350_MeanScore['Mean Score']= pd.to_numeric(df_13K350_MeanScore)
df_13K350_MeanScore['Mean Score'].dtype
```

#Average of the Mean Score for school 13K350 High School Common Core Algebra Exam


```python
df_13K350_mean = df_13K350_MeanScore['Mean Score'].mean()
display(df_13K350_mean)
```

#Average of the Mean Score for Entire Brooklyn Borough for the High School Common Core Algebra Exam

```python
df_B_HS_CCA_MeanScore = df_B_HS_CCA['Mean Score']
df_B_HS_CCA_MeanScore['Mean Score']= pd.to_numeric(df_B_HS_CCA_MeanScore)
df_B_HS_CCA_MeanScore['Mean Score'].dtype
```

#Average of the Mean Scores for all High Schools in Brooklyn for Common Core Algebra Exam

```python
df_B_HS_CCA_mean = df_B_HS_CCA_MeanScore['Mean Score'].mean()
display(df_B_HS_CCA_mean)
```

```python
df_Q_HS_CCA_MeanScore = df_Q_HS_CCA['Mean Score']
df_Q_HS_CCA_MeanScore['Mean Score']= pd.to_numeric(df_Q_HS_CCA_MeanScore)
df_Q_HS_CCA_MeanScore['Mean Score'].dtype
```

#Average of the Mean Scores for all High Schools in Queens for Common Core Algebra Exam

```python
df_Q_HS_CCA_mean = df_Q_HS_CCA_MeanScore['Mean Score'].mean()
display(df_Q_HS_CCA_mean)
```

#Create new Data Frame with Data Findings to assist with Data Viz chart

```python
Avg_CCA_Scores = pd.DataFrame([('Urban Assembly School for Music and Art',df_13K350_mean),('Brooklyn High Schools', df_B_HS_CCA_mean),('Queens High Schools', df_Q_HS_CCA_mean)], columns=('High School Groupings','Average of Mean Scores'))
display(Avg_CCA_Scores)
```

```python
Avg_CCA_Scores.plot(kind= 'bar', x='High School Groupings', y='Average of Mean Scores', ylim=(50,70), color=['red','green','orange'], edgecolor= 'black', title= 'Average of Mean Score Common Core Algebra Scores', grid=True, ylabel = 'Average Scores')
```

```python
df_Q_HS_CCA_max = df_Q_HS_CCA_MeanScore['Mean Score'].max()
display(df_Q_HS_CCA_max)
```

```python
df_B_HS_CCA_max = df_B_HS_CCA_MeanScore['Mean Score'].max()
display(df_B_HS_CCA_max)
```

```python
df_13K350_max= df_13K350_MeanScore['Mean Score'].max()
display(df_13K350_max)
```

```python

Max_CCA_Scores = pd.DataFrame([('Urban Assembly School for Music and Art',df_13K350_max),('Brooklyn High Schools', df_B_HS_CCA_max),('Queens High Schools', df_Q_HS_CCA_max)], columns=('High School Groupings','Max of Mean Scores'))
display(Max_CCA_Scores)
```

```python
Max_CCA_Scores.plot(kind= 'bar', x='High School Groupings', y='Max of Mean Scores', ylim=(60,100), color=['red','green','orange'], edgecolor= 'black', title= 'Max of Mean Score Common Core Algebra Scores', grid=True, ylabel = 'Max Scores')
```
