##Clean data by converting invalid data as missing data

```python
missing_values = ["?", "Unknown/Invalid"]
df1 = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv', na_values = missing_values)
display(df1)
```

##Limit Columns to only show race, gender, lenght of hospital stay, and insulin level

```python
df_hospital_stay_limited = df1[['encounter_id','age','race','gender','time_in_hospital','insulin']]
display(df_hospital_stay_limited)
```

```python
df_race_groups= df_hospital_stay_limited.groupby(['race'])['encounter_id'].count()
display(df_race_groups)
```

```python
df_hospital_stay_limited.groupby(['race','time_in_hospital'])['encounter_id'].count()
```

```python
df_hospital_stay_limited.loc[df['race'].str.contains('AfricanAmerican')]
```

```python
df_average_length_by_race = df_hospital_stay_limited.groupby('race')['time_in_hospital'].mean()
print(df_average_length_by_race)
```

```python
df_max_length_by_race = df_hospital_stay_limited.groupby('race')['time_in_hospital'].max()
print(df_max_length_by_race)
```

```python
df_sum_length_by_race = df_hospital_stay_limited.groupby('race')['time_in_hospital'].sum()
print(df_sum_length_by_race)
```

```python
df_sum_length_by_race.plot(x='race', y='time_in_hospital', kind='bar', color='purple')
plt.xlabel('Race')
plt.ylabel('Days in Hospital')
plt.title('Total Sum of Hospital Stay by Race')
```

```python
df_average_length_by_race.plot(x='race', y='time_in_hospital', kind='bar', color='blue')
plt.xlabel('Race')
plt.ylabel('Days in Hospital')
plt.title('Average Days of Hospital Stay by Race')
```

```python
df_median_length_by_race = df_hospital_stay_limited.groupby('race')['time_in_hospital'].median()
print(df_median_length_by_race)
```

```python
df_average_length_by_age = df_hospital_stay_limited.groupby('age')['time_in_hospital'].mean()
print(df_average_length_by_age)
```

```python
df_average_length_by_age.plot(x='age', y='time_in_hospital', kind='bar', color='pink')
plt.xlabel('Age')
plt.ylabel('Days in Hospital')
plt.title('Average Days of Hospital Stay by Age Groups')
```

```python
df_median_length_by_age = df_hospital_stay_limited.groupby('age')['time_in_hospital'].median()
print(df_median_length_by_age)
```

```python
df_sum_length_by_age = df_hospital_stay_limited.groupby('age')['time_in_hospital'].sum()
print(df_sum_length_by_age)
```

```python
df_sum_length_by_age.plot(x='age', y='time_in_hospital', kind='bar', color='blue')
plt.xlabel('Age')
plt.ylabel('Days in Hospital')
plt.title('Total Days of Hospital Stay by Age Groups')
```

##Conclusions:
1. Caucasians had the largest sum of stays at the hospital but African Americans had longer hospital stays on average
2. Amongst all races, people aged 80-90 had the most hospital stays and had lobger hospital stays on average
